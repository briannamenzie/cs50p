### **While Loops**
- The `while` loop is nearly universal throughout all coding languages
- Such a loop will repeat a block of code over and over again
- The below code will execute `print("meow")` multiple times, and it will never stop (it will loop forever). 
- `while` loop works by repeatedly asking if the condition of the loop has been fulfilled.
    - In this case, the interpreter is asking "does `i` not equal zero?"
    - When you get stuck in a loop that executes forever, you can press `Ctrl+C` on your keyboard to break out of the loop

```python
# i = 3
# while i != 0:
#     print("meow")
```

- To fix this loop, we can edit our code as follows:
- Notice that now, the i reduces by 1 for each "iteration" of the loop.
- By iteration, we mean one cycle through the loop. 
    - The first iteration is the 0th iteration
    - Same counting applies to indexing and all else in programming

```python
# i = 3
# while i != 0:
#   print("meow")
#   i = i - 1
```

- To further improve the loop, we can code i=i+1, which assigns the value of i to the left.

- If you execute the code above, you'll see it meows three times. However, we can still further improve, because it's best practice in programming to begin counting with zero.

```python
# # starting at i=1, while i is less than or equal to 3, print "meow" and then add 1 to i
# i = 1
# while i <= 3:
#     print("meow")
#     i = i + 1

# # ************************

# # starting at i=0 (best practice in programming), while i is less than 3, print "meow" and then add 1 to i
# i = 0
# while i < 3:
#     print("meow")
#     i += 1 # same as saying i = i + 1
```

### **For Loops**

- A `for` loop iterates through a list of items.
- Typically way cleaner than the `while` loop
- See below as an example:

```python
# for i in [0, 1, 2]:
#     print("meow")
```

- In the above code, `i` begins with 0, meows, `i` is assigned to 1, meows, then finally, `i` is assigned 2, meows, then the loop ends.
- Can improve isomg the `range` function:

```python
# for i in range(3):
#     print("meow")
```

- Notice how `range(3)` provides back three values (0, 1, 2) automatically. So if I did `range(10)`, it would count from 0 to 9.
- Also note that the `i` is the place where the number of the iteration of the loop gets stored. We can replace `i` with whatever we'd like, as it doesn't really have significance in our code.
- Can further change/improve the way we print "meow" three times...

```python
# # will print "meow" three times, all on the same line, because the end="" argument prevents print from adding a newline after each meow
# # output is meowmeowmeow
# print("meow" * 3)

# # OR 

# # produces three meows, each on its own line
# # the end="" argument prevents print from adding a newline after each meow
# # the "\n" adds a newline after each meow, so that they are printed on separate lines
# print("meow\n" * 3, end="")
```

### **Improving with User Input**

- We can use loops as a way of validating the input of the user
- A common paradigm within Python is to use a while loop to validate the input of the user
- For example, we can try prompting the user for a number greater than or equal to 0:

```python
# while True:
#     n = int(input("What's n? " ))
#     if n<0:
#         continue
#     else:
#         break

# # continue: goes back to the top of the loop and starts the next iteration, skipping any code that follows it in the current iteration
# # break: exits the loop immediately, regardless of the loop's condition
```

- Notice that we’ve introduced two new keywords in Python, `continue` and `break`. 
    1) `continue` explicitly tells Python to go to the next iteration of a loop. 
    2) `break`, on the other hand, tells Python to “break out” of a loop early before it has finished all of its iterations. 
- In this case, we’ll continue to the next iteration of the loop when `n` is less than 0—ultimately reprompting the user with “What’s n?”. 
- If, though, `n` is greater than or equal to 0, we’ll `break` out of the loop and allow the rest of our program to run.
- It turns out that continue keyword is redundant in this case. We can improve our code as follows:

```python
## this code will keep asking the user for a positive integer until they provide one, and then it will print "meow" that many times

# while True:
#     n = int(input("What's n? "))
#     if n > 0:
#         break

# for _ in range(n):
#     print("meow")
```
